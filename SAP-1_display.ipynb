# SAP-1 Emulator
### SImulation of a led strip using Matplolib

```python
import matplotlib.pyplot as plt

registers = {'Register A': 0b100, 'ALU' : 23, 'Register B': 0b0, 'OUT': 0b0}
```

```python
def plot_display():
  fig, ax = plt.subplots()
  fig.set_size_inches(6,3)
  plt.title('SAP-1 Display', fontsize= 20)
  x = range(8)
  y = 0
  ax.set_facecolor('mediumseagreen')
  ax.set_xticks([])
  ax.set_yticks(range(len(registers)) , labels = list(registers.keys()), fontsize = 15)
  ax.invert_yaxis() #Invert the y axis so the first value is above
  for i in registers:
    bits = get_bits(registers[i]) #Get the single bits
    colors = ['#600000' if b == 0 else '#ff2020' for b in bits] 
    ax.scatter(x,[y]*8, c = colors, edgecolors='k', s=300)
    ax.text(10, y, f'{registers[i]:08b}', fontsize = 15, verticalalignment='center')
    y = y + 1 #Move a line up
  ax.margins(0.3)
    
  

def get_bits(a):
  bits = [*f'{a:08b}']
  return [int(b) for b in bits]
```

```python
print(len(registers))
```

```python
plot_display()
```
